 # 내가 푼 거

```python
#k+1이 분모 분자 합임


def return_k(input_num):
    for k in range(1,10000):
        sum_val = 0
        for i in range (k+1):
            sum_val += i
            # print("k:", k, "sum:", sum_val, "i:", i)
            if input_num >= sum_val:
                k_candidate = k
            if input_num <= sum_val:
                return k_candidate, sum_val



input_num=int(input())

k,sum_val=return_k(input_num)
# print(sum_val)

diff=sum_val-input_num
# print("diff:", diff)

if k%2 == 0:
    # print("k is even")
    numerator = k-diff
    denominator = diff+1
    print(str(numerator)+"/"+str(denominator))
if k%2 == 1:
    # print("k is odd")
    numerator = diff+1
    denominator = k-diff
    print(str(numerator)+"/"+str(denominator))
```
