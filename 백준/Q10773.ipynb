import sys

# K     readline 
K = int(sys.stdin.readline())
stack = []

for _ in range(K):
    num = int(sys.stdin.readline())
    
    # 0    (push)
    if num != 0:
        stack.append(num)
    # 0   (pop)
    #        ,
    #  if stack:      .
    else:
        stack.pop()

print(sum(stack))
