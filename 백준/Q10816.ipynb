### \#\# Gemini  

```python
import sys
from collections import Counter

# N       ,
#       .
_ = sys.stdin.readline()

#      Counter  .
# map(int, ...)     .
card_counts = Counter(map(int, sys.stdin.readline().split()))

_ = sys.stdin.readline()
targets = list(map(int, sys.stdin.readline().split()))

#    .
answers = []
for target in targets:
    # Counter target    .
    #    0  Counter  .
    answers.append(card_counts[target])

#  '   '    .
# map(str, ...)       join.
print(' '.join(map(str, answers)))
```
