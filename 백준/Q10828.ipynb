````python
import sys

lst=[]
N=int(input())

for _ in range(N):
    temp_lst=[]
    temp_lst=sys.stdin.readline().split()

    if len(temp_lst)==2:
        order,val=temp_lst
    else:
        order=temp_lst[0]

    if order=='push':
        lst.append(int(val))


    elif order=='pop':
        if len(lst)!=0:
            print(lst.pop())
        elif len(lst)==0:
            print('-1')

    elif order=='size':
        print(len(lst))

    elif order=='empty':
        if len(lst)==0:
            print('1')
        else:
            print('0')

    elif order=='top':
        if len(lst)!=0:
            print(lst[len(lst)-1])
        else:
            print('-1')
````

```python
import sys

# 1. N   sys.stdin.readline() 
N = int(sys.stdin.readline())
stack = []

for _ in range(N):
    command = sys.stdin.readline().split()
    order = command[0]

    if order == 'push':
        stack.append(command[1])
    
    elif order == 'pop':
        # 2. if stack:    
        if stack:
            print(stack.pop())
        else:
            print(-1)
    
    elif order == 'size':
        print(len(stack))
    
    elif order == 'empty':
        print(0 if stack else 1) # 3.     
    
    elif order == 'top':
        # 4. stack[-1]   
        if stack:
            print(stack[-1])
        else:
            print(-1)
```
