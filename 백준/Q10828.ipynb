````python
import sys

lst=[]
N=int(input())

for _ in range(N):
    temp_lst=[]
    temp_lst=sys.stdin.readline().split()

    if len(temp_lst)==2:
        order,val=temp_lst
    else:
        order=temp_lst[0]

    if order=='push':
        lst.append(int(val))


    elif order=='pop':
        if len(lst)!=0:
            print(lst.pop())
        elif len(lst)==0:
            print('-1')

    elif order=='size':
        print(len(lst))

    elif order=='empty':
        if len(lst)==0:
            print('1')
        else:
            print('0')

    elif order=='top':
        if len(lst)!=0:
            print(lst[len(lst)-1])
        else:
            print('-1')
````

### \#\# Gemini 추천 코드

```python
import sys

# 1. N을 입력받는 부분도 sys.stdin.readline()으로 변경
N = int(sys.stdin.readline())
stack = []

for _ in range(N):
    command = sys.stdin.readline().split()
    order = command[0]

    if order == 'push':
        stack.append(command[1])
    
    elif order == 'pop':
        # 2. if stack: 으로 더 간결하게 표현
        if stack:
            print(stack.pop())
        else:
            print(-1)
    
    elif order == 'size':
        print(len(stack))
    
    elif order == 'empty':
        print(0 if stack else 1) # 3. 삼항 연산자로 한 줄로 표현
    
    elif order == 'top':
        # 4. stack[-1]로 더 간결하게 표현
        if stack:
            print(stack[-1])
        else:
            print(-1)
```

### \#\# 2. 파이썬 문법 측면 (Pythonic Style)

#### **내 풀이에서 개선할 점 (매우 사소한 부분)**

1.  **마지막 남은 `input()`:** `for` 루프 안은 잘 수정하셨지만, 맨 처음 `N=int(input())` 부분이 아직 남아있습니다. 이 부분도 `N = int(sys.stdin.readline())`으로 바꿔주면 모든 입출력이 최적화됩니다.
2.  **빈 리스트 확인:** `if len(lst) != 0:` 보다는, 파이썬에서는 **`if lst:`** 라고 쓰는 것이 훨씬 더 일반적이고 간결합니다. (비어있지 않은 모든 컨테이너는 `True`로 취급됩니다.)
3.  **Top 요소 접근:** `lst[len(lst)-1]` 보다는, 리스트의 마지막 요소를 가리키는 \*\*`lst[-1]`\*\*을 사용하는 것이 훨씬 더 간결하고 Pythonic합니다.

### \#\# 3. 또 다른 측면: 코드의 일관성과 단순성

이번에 칭찬하고 싶은 부분은 **코드의 단순화**입니다. 이전 코드에 있던 불필요한 `push` 함수를 제거하고, `lst.append()`를 직접 호출하는 방식으로 변경하셨습니다.

  * **왜 좋은 변화인가?**
      * 이 문제에서 `push` 함수는 단순히 `append`를 한 번 더 감싸는 역할만 했습니다. 이런 간단한 기능은 굳이 함수로 분리하는 것보다 직접 호출하는 것이 코드의 흐름을 더 명확하게 보여줍니다.
      * 불필요한 `temp_lst` 변수를 매번 초기화할 필요 없이, `command = sys.stdin.readline().split()` 처럼 바로 변수에 할당하는 것이 더 깔끔합니다.

이처럼 \*\*'꼭 필요한가?'\*\*를 스스로 질문하며 코드를 더 단순하고 직접적으로 만드는 과정은 매우 훌륭한 습관입니다. 이번 코드는 기능적으로 완벽하며, 여기서 더 나아간다면 코드를 '얼마나 더 파이썬답게 다듬는가'의 문제입니다. 정말 잘하셨습니다\!
