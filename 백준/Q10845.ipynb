```python
from collections import deque
import sys

dq=deque()

for _ in range(int(input())):

    lst=sys.stdin.readline().split()

    if len(lst) > 1 :
        dq.append(lst[1])

    elif lst[0]=='pop':
        if dq:
            print(dq.popleft())
        else:
            print(-1)

    elif lst[0]=='size':
        print(len(dq))

    elif lst[0]=='empty':
        if dq:
            print(0)
        else:
            print(1)

    elif lst[0]=='front':
        if dq:
            print(dq[0])
        else:
            print(-1)

    elif lst[0]=='back':
        if dq:
            print(dq[-1])
        else:
            print(-1)        

```
