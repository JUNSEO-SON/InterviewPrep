```python
import sys

#input = sys.stdin.readline

N=int(input())
nums=list(map(int,input().split()))
lis=[1 for _ in range(N)]

for i in range(N):
    for j in range(i):
        if nums[j]<nums[i]:
            lis[i]=max(lis[j]+1,lis[i])

print(max(lis))
```
