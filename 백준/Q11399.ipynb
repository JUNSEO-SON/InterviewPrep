```python
import sys

input = sys.stdin.readline
N = int(input())

#   .
times = sorted(map(int, input().split()))

#    (p_i) (N-i)    .
# p_1 N, p_2 N-1, ..., p_N 1 .
total_time = 0
for i in range(N):
    total_time += times[i] * (N - i)

print(total_time)

#    (Pythonic)
```
