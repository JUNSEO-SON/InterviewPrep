```python
N =int(input())

list_xy = []

for _ in range(N):
    x,y = map(int,input().split())
    list_xy.append((x,y))
    

sorted_list_xy = sorted(list_xy, key=lambda x: (x[0],x[1]))


for tup in sorted_list_xy:
    print(*tup)
```

```python
import sys

# 빠른 입력을 위해 sys.stdin.readline 사용
N = int(sys.stdin.readline())

# 리스트 컴프리헨션을 사용하여 N개의 좌표를 한 번에 입력받음
coords = [tuple(map(int, sys.stdin.readline().split())) for _ in range(N)]

# 리스트를 정렬. 튜플의 기본 정렬 방식이 문제의 요구사항과 동일하므로 key가 필요 없음
coords.sort()

# 정렬된 리스트를 순회하며 출력
for x, y in coords:
    print(x, y)




```

### \#\# 1. 알고리즘 측면

  * **개선할 점:** N이 100,000개까지 주어지는 문제에서는 `input()` 함수가 너무 느려 시간 초과가 발생할 수 있습니다. 이는 알고리즘 자체의 문제가 아니라 입력을 받아오는 과정의 병목 현상 때문입니다. 경쟁 프로그래밍에서는 대량의 입력을 받을 때 `sys.stdin.readline()`을 사용하는 것이 거의 표준처럼 여겨집니다.

### \#\# 2. 파이썬 문법 측면 (Pythonic Style)

#### **내 풀이에서 개선할 점**

1.  **정렬 키(key)의 중복:** `key=lambda x: (x[0], x[1])`은 정확한 코드이지만, 사실 이 부분은 **불필요**합니다. 파이썬은 리스트에 튜플이 여러 개 있을 경우, 기본적으로 튜플의 첫 번째 요소를 기준으로 정렬하고, 그 값이 같으면 두 번째 요소를 기준으로 정렬합니다. 즉, 문제의 요구사항과 파이썬의 **기본 튜플 정렬 방식이 완벽하게 일치**하므로 `key` 인자를 생략해도 동일한 결과가 나옵니다.

#### **Gemini 풀이에서 배울 점 (Pythonic)**

1.  **빠른 입출력:** `import sys` 후 `sys.stdin.readline()`을 사용하여 실행 속도를 크게 향상시켰습니다.
2.  **리스트 컴프리헨션(List Comprehension):** `for` 루프를 사용해 `list.append()`를 반복하는 대신, 리스트 컴프리헨션을 사용하면 단 한 줄로 더 간결하고 빠르게 입력 데이터를 리스트로 만들 수 있습니다.
3.  **내장 기능 활용:** 위에서 설명했듯이, 별도의 `key` 없이 `coords.sort()`나 `sorted(coords)`를 호출하여 파이썬의 강력한 내장 정렬 기능을 최대한 활용했습니다.
      * (`coords.sort()`는 리스트 자체를 정렬(in-place)하여 메모리를 조금 더 효율적으로 사용하고, `sorted(coords)`는 정렬된 새로운 리스트를 반환합니다.)
