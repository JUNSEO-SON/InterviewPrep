```python
N =int(input())

list_xy = []

for _ in range(N):
    x,y = map(int,input().split())
    list_xy.append((x,y))
    

sorted_list_xy = sorted(list_xy, key=lambda x: (x[0],x[1]))


for tup in sorted_list_xy:
    print(*tup)
```

```python
import sys

#    sys.stdin.readline 
N = int(sys.stdin.readline())

#    N    
coords = [tuple(map(int, sys.stdin.readline().split())) for _ in range(N)]

#  .        key  
coords.sort()

#    
for x, y in coords:
    print(x, y)




```
