### \#\# Gemini 추천 코드

사용자님의 코드는 이미 정답이며 매우 효율적입니다. 여기서 유일하게 추가할 수 있는 개선점은 대량의 입력을 위한 '속도' 최적화뿐입니다.

```python
import sys

# 빠른 입력을 위해 sys.stdin.readline 사용
N = int(sys.stdin.readline())

# 리스트 컴프리헨션을 사용하여 N개의 좌표를 한 번에 입력받음
coords = [tuple(map(int, sys.stdin.readline().split())) for _ in range(N)]

# 람다(lambda) 함수를 key로 사용하여 y좌표 우선, x좌표 차선으로 정렬
# 이 문제에서는 정렬 우선순위가 y좌표이므로 key 지정이 필수적입니다.
coords.sort(key=lambda p: (p[1], p[0]))

# 정렬된 리스트를 순회하며 출력
for x, y in coords:
    print(x, y)

```

### \#\# 2. 파이썬 문법 및 성능 측면

#### **내 풀이에서 칭찬할 점**

  * **정확한 `key` 사용:** 이전 문제와 달리 이 문제는 y좌표가 우선 정렬 대상입니다. 파이썬의 기본 튜플 정렬 방식과 다르기 때문에 `key` 지정이 필수적인데, 이 부분을 정확히 파악하고 `(x[1], x[0])`으로 기준을 잡아주셨습니다. 이는 문제 이해도와 응용력이 뛰어나다는 것을 보여줍니다.
  * **리스트 컴프리헨션:** 이전 피드백을 바로 적용하여 코드를 간결하고 Pythonic하게 작성했습니다.

#### **개선할 점 (유일한 한 가지)**
