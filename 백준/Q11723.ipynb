````python
import sys
a=set()

M=int(sys.stdin.readline())

for _ in range(M):
    temp_lst=[]
    temp_lst=sys.stdin.readline().split()

    if len(temp_lst)==2:
        order,val=temp_lst
        val=int(val)
    else:
        order=temp_lst[0]

    if order=='add':
        a.add(int(val))

    elif order=='check':
        if val in a:
            print(1)
        else:
            print(0)

    elif order=='remove':
        a.discard(val)

    elif order=='empty':
            a.clear()

    elif order=='all':
        a=set([x for x in range(1,21)])

    elif order=='toggle':
        if val in a:
            a.remove(val)
        else:
            a.add(val)

````

        print(1 if int(command[1]) in s else 0)
    elif command[0] == "toggle":

        if x in s:
            s.remove(x)
        else:
            s.add(x)
    elif command[0] == "all":
        s = set(range(1, 21))
    elif command[0] == "empty":
        s.clear()
