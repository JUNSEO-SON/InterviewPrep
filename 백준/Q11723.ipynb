````python
import sys
a=set()

M=int(sys.stdin.readline())

for _ in range(M):
    temp_lst=[]
    temp_lst=sys.stdin.readline().split()

    if len(temp_lst)==2:
        order,val=temp_lst
        val=int(val)
    else:
        order=temp_lst[0]

    if order=='add':
        a.add(int(val))

    elif order=='check':
        if val in a:
            print(1)
        else:
            print(0)

    elif order=='remove':
        a.discard(val)

    elif order=='empty':
            a.clear()

    elif order=='all':
        a=set([x for x in range(1,21)])

    elif order=='toggle':
        if val in a:
            a.remove(val)
        else:
            a.add(val)

````

### 1. 알고리즘 측면: ★★★★☆ (4점)

**장점:**
- **주어진 연산을 모두 구현**: add, remove, check, toggle, all, empty 등 모든 연산을 올바르게 처리하고 있습니다.
- **시간복잡도**: 각 연산의 시간 복잡도는 O(1)으로 최적화되어 있습니다. 특히 `set` 자료구조를 사용하여 효율적인 연산을 수행하고 있습니다.
- **입력 처리**: `sys.stdin.readline()`을 사용하여 입력 속도를 크게 개선했습니다.

**개선 필요:**
- **temp_lst 초기화 불필요**: `temp_lst = []`를 선언한 뒤 바로 덮어씌우는 부분은 불필요합니다. 이를 제거하면 코드가 더 간결해집니다.
- **문자열 처리**: `command = input().strip().split()`로 입력을 처리하는 것이 더 안전합니다. 현재 코드에서는 공백이나 줄 바꿈 문자가 남아 있을 가능성이 있습니다.
- **toggle에서 중복 remove**: `toggle` 연산에서 `a.remove(val)`를 사용하고 있지만, 이미 `discard()`를 사용하고 있으므로 일관성을 위해 `discard()`로 통일하는 것이 좋습니다.
