 # 내가 푼 거

```python
#k+1이 분모 분자 합임


def return_k(input_num):
    for k in range(1,10000):
        sum_val = 0
        for i in range (k+1):
            sum_val += i
            # print("k:", k, "sum:", sum_val, "i:", i)
            if input_num >= sum_val:
                k_candidate = k
            if input_num <= sum_val:
                return k_candidate, sum_val



input_num=int(input())

k,sum_val=return_k(input_num)
# print(sum_val)

diff=sum_val-input_num
# print("diff:", diff)

if k%2 == 0:
    # print("k is even")
    numerator = k-diff
    denominator = diff+1
    print(str(numerator)+"/"+str(denominator))
if k%2 == 1:
    # print("k is odd")
    numerator = diff+1
    denominator = k-diff
    print(str(numerator)+"/"+str(denominator))
```

# Gemini가 풀어준거

```python
# 사용자로부터 X 값을 입력받습니다.
X = int(input())

# X가 몇 번째 대각선 라인에 있는지 찾습니다.
line = 1
while X > line:
    X -= line
    line += 1

# 대각선 라인(line)이 짝수일 때와 홀수일 때를 구분합니다.
if line % 2 == 0:  # 짝수 라인일 경우 (분자 증가, 분모 감소)
    numerator = X
    denominator = line - X + 1
else:  # 홀수 라인일 경우 (분자 감소, 분모 증가)
    numerator = line - X + 1
    denominator = X

# 결과를 "분자/분모" 형태로 출력합니다.
print(f"{numerator}/{denominator}")
```
