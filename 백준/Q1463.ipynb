````python
import sys
input=sys.stdin.readline

N=int(input())

dp = [float('inf')] * (N + 1)  
dp[1]=0

for i in range(2,N+1):
    if i%3==0:
        type1=dp[i//3]+1
    else:
        type1 = float('inf')
    if i%2==0:
        type2=dp[i//2]+1
    else:
        type2 = float('inf')

    type3=dp[i-1]+1

    dp[i]=min(type1,type2,type3)

print(dp[N])
````

```python
import sys

N = int(sys.stdin.readline())

# dp[i] i 1    
dp = [0] * (N + 1)

# 2 N DP  
for i in range(2, N + 1):
    # 1. , i i-1 +1     
    dp[i] = dp[i-1] + 1
    
    # 2.  i 2  , /2        
    if i % 2 == 0:
        dp[i] = min(dp[i], dp[i//2] + 1)
        
    # 3.  i 3  , /3         
    if i % 3 == 0:
        dp[i] = min(dp[i], dp[i//3] + 1)

print(dp[N])
```
