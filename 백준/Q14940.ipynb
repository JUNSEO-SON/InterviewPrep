```python
import sys
from collections import deque

input = sys.stdin.readline

n, m = map(int, input().split())
original_map = []
start_y, start_x = -1, -1

for i in range(n):
    row = list(map(int, input().split()))
    original_map.append(row)
    if 2 in row:
        start_y = i
        start_x = row.index(2)

#       
distance_map = [[0] * m for _ in range(n)]
for y in range(n):
    for x in range(m):
        if original_map[y][x] == 1:
            distance_map[y][x] = -1 #     -1 

def bfs(start_y, start_x):
    dq = deque()
    #  (y, x)   
    dq.append((start_y, start_x))
    #   0  (   )
    distance_map[start_y][start_x] = 0

    dy = [-1, 1, 0, 0]
    dx = [0, 0, -1, 1]

    while dq:
        #   
        y, x = dq.popleft()
        
        #    
        for i in range(4):
            ny = y + dy[i]
            nx = x + dx[i]
            
            #   ,    ( -1)
            if 0 <= ny < n and 0 <= nx < m and distance_map[ny][nx] == -1:
                #     ,
                dq.append((ny, nx))
                # "   + 1"   ( )
                distance_map[ny][nx] = distance_map[y][x] + 1

bfs(start_y, start_x)

#   
for row in distance_map:
    print(*row)
```
