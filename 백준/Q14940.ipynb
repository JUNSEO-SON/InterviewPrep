````python
import sys
from collections import deque

input = sys.stdin.readline

n,m=map(int,input().split())
listmap=[]
for i in range(n):
    temp=list(map(int,input().split()))
    listmap.append(temp)
    for idx,val in enumerate(temp):
        if val==2:
            start_y=i
            start_x=idx

dy=[-1,1,0,0]
dx=[0,0,-1,1]
countmap=[[0 for _ in range(m)] for _ in range(n)]

for y in range(n):
    for x in range(m):
        if listmap[y][x]==1:
            countmap[y][x]=-1

def bfs(y,x):
    dq=deque()
    count=0
    dq.append((y,x,count))

    while(dq):
        y,x,count=dq.popleft()
        count+=1
        for i in range(4):
            nx=x+dx[i]
            ny=y+dy[i]
            if 0<=nx<m and 0<=ny<n:
                if listmap[ny][nx]==1:
                    countmap[ny][nx]=count
                    listmap[ny][nx]=3
                    dq.append((ny,nx,count))


bfs(start_y,start_x)
for comp in countmap:
    print(*comp)

````

### \#\# Gemini 추천 코드 (표준 BFS 구현)

```python
import sys
from collections import deque

input = sys.stdin.readline

n, m = map(int, input().split())
original_map = []
start_y, start_x = -1, -1

for i in range(n):
    row = list(map(int, input().split()))
    original_map.append(row)
    if 2 in row:
        start_y = i
        start_x = row.index(2)

# 방문 여부와 거리를 한 번에 기록할 배열
distance_map = [[0] * m for _ in range(n)]
for y in range(n):
    for x in range(m):
        if original_map[y][x] == 1:
            distance_map[y][x] = -1 # 갈 수 있는 땅은 -1로 초기화

def bfs(start_y, start_x):
    dq = deque()
    # 큐에는 (y좌표, x좌표) 위치 정보만 저장
    dq.append((start_y, start_x))
    # 시작점의 거리는 0으로 설정 (이것이 방문 처리의 시작)
    distance_map[start_y][start_x] = 0

    dy = [-1, 1, 0, 0]
    dx = [0, 0, -1, 1]

    while dq:
        # 현재 위치를 꺼냄
        y, x = dq.popleft()
        
        # 네 방향의 이웃을 확인
        for i in range(4):
            ny = y + dy[i]
            nx = x + dx[i]
            
            # 지도 범위 안이고, 아직 방문하지 않은 땅(거리가 -1)이라면
            if 0 <= ny < n and 0 <= nx < m and distance_map[ny][nx] == -1:
                # 큐에 다음 탐색 위치를 추가하고,
                dq.append((ny, nx))
                # "이전 노드의 거리 + 1"로 거리를 기록 (방문 처리)
                distance_map[ny][nx] = distance_map[y][x] + 1

bfs(start_y, start_x)

# 최종 결과 출력
for row in distance_map:
    print(*row)
```

### \#\# 1. 알고리즘 측면: 왜 이 코드가 통과했을까?

이것이 오늘 배울 가장 중요한 부분입니다. 사용자님의 코드는 BFS의 거리 계산 방식에 논리적인 허점이 있지만, **이 문제의 특성과 테스트 케이스 덕분에** 운 좋게 통과할 수 있었습니다.

  * **사용자님의 거리 계산:** `count`가 큐에서 노드를 **꺼낼 때마다 1씩 증가**합니다.
  * **표준 BFS 거리 계산:** 거리는 항상 **`부모 노드의 거리 + 1`** 입니다.

**왜 통과했을까?**
이 문제의 지도는 대부분 넓게 퍼지는 단순한 형태일 가능성이 높습니다. 이런 경우, 큐에 들어간 순서와 레벨(거리) 순서가 우연히 거의 일치하게 되어, `count`가 증가하는 방식이 결과적으로 올바른 거리처럼 보이게 된 것입니다. 만약 여러 경로가 한 점으로 모이거나, 매우 복잡하게 얽힌 미로 같은 테스트 케이스가 있었다면, 아마 틀렸을 가능성이 높습니다.

### \#\# 3. 또 다른 측면: 방문 처리(Visit Check)의 시점

  * **사용자님의 코드:** `listmap[ny][nx]==1`인지 확인하고, 큐에 넣은 뒤, `listmap[ny][nx]=3`으로 방문 처리합니다.
  * **추천 코드:** `distance_map[ny][nx] == -1`인지 확인하고, **큐에 넣기 직전에** `distance_map[ny][nx] = ...`으로 방문 처리(거리 기록)를 합니다.

**왜 큐에 넣기 직전에 하는 것이 더 안전할까?**
큐에 넣은 '후'에 방문 처리를 하면, 어떤 노드가 큐에 이미 예약되어 있음에도 불구하고, 다른 경로를 통해 또다시 큐에 중복으로 들어가는 것을 막을 수 없습니다. 이는 비효율을 유발하고, 복잡한 그래프에서는 논리적 오류의 원인이 될 수 있습니다. \*\*"큐에 넣기로 결정하는 순간, 바로 방문했다고 도장을 찍는 것"\*\*이 가장 안전하고 표준적인 방법입니다.
