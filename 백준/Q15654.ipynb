```python
import sys

#input = sys.stdin.readline
N,M=map(int,input().split())
nums=list(map(int,input().split()))
nums.sort()
ans=[]
temp=[]
visited=[False for _ in range(N)]

def select():
    if len(temp)==M:
        ans.append(temp.copy())
        return
    
    for i in range(N):
        if not visited[i]:
            visited[i]=True
            temp.append(nums[i])
            select()
            temp.pop()
            visited[i]=False


select()

for a in ans:
    print(*a)
```
