```python
import sys

input = sys.stdin.readline

N, M = map(int, input().split())
pokemon_dictnum = {}
pokemon_dictname = {}

#   
for i in range(1, N + 1):
    pokemon_name = input().strip()
    pokemon_dictnum[i] = pokemon_name
    pokemon_dictname[pokemon_name] = i

#  
for _ in range(M):
    query = input().strip()

    #  
    if query.isdigit():
        try:
            num = int(query)
            if 1 <= num <= N:
                print(pokemon_dictnum[num])
            else:
                print("Invalid number")
        except ValueError:
            print("Invalid input")

    #  
    elif query.isalpha():
        try:
            print(pokemon_dictname[query])
        except KeyError:
            print("Pokemon not found")

    #   
    else:
        print("Invalid input")
```
