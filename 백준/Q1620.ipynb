````python
import sys

input=sys.stdin.readline

N,M=map(int,input().split())
pokemon_dictnum={}
pokemon_dictname={}

for i in range(1,N+1):
    pokename=input().strip()
    pokemon_dictnum[i]=pokename
    pokemon_dictname[pokename]=i


for _ in range(M):
    query=input().strip()
    if query.isalpha():
        print(pokemon_dictname[query])
    else:
        print(pokemon_dictnum[int(query)])
````

### 수정된 코드 예시

```python
import sys

input = sys.stdin.readline

N, M = map(int, input().split())
pokemon_dictnum = {}
pokemon_dictname = {}

# 포켓몬 정보 입력
for i in range(1, N + 1):
    pokemon_name = input().strip()
    pokemon_dictnum[i] = pokemon_name
    pokemon_dictname[pokemon_name] = i

# 질의 처리
for _ in range(M):
    query = input().strip()

    # 숫자인 경우
    if query.isdigit():
        try:
            num = int(query)
            if 1 <= num <= N:
                print(pokemon_dictnum[num])
            else:
                print("Invalid number")
        except ValueError:
            print("Invalid input")

    # 문자열인 경우
    elif query.isalpha():
        try:
            print(pokemon_dictname[query])
        except KeyError:
            print("Pokemon not found")

    # 잘못된 입력인 경우
    else:
        print("Invalid input")
```
