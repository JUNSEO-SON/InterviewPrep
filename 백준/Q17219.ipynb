```python
import sys

# input  sys.stdin.readline    .
input = sys.stdin.readline

N, M = map(int, input().split())

# (key) (value)  
site_to_password = {}

# N     
for _ in range(N):
    # split()     (unpacking)  .
    site, password = input().split()
    site_to_password[site] = password

# M     
for _ in range(M):
    # readline() (\n)  .strip()   .
    site_to_find = input().strip()
    print(site_to_password[site_to_find])
```
