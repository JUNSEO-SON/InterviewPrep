````python
import sys

input=sys.stdin.readline

N,M=map(int,input().split())

never_heard=set([input().strip() for _ in range(N)])
never_seenheard=[]

for _ in range(M):
    never_seen=input().strip()
    if never_seen in never_heard:
        never_seenheard.append(never_seen)

never_seenheard.sort()

print(len(never_seenheard))
print(*never_seenheard,sep='\n')
````

### 모범 답안 (참고용)

```python
import sys

input = sys.stdin.readline

N, M = map(int, input().split())
never_heard = set(input().strip() for _ in range(N))
never_seen = [input().strip() for _ in range(M)]

# 교집합 구하기
intersection = sorted(never_heard & set(never_seen))

print(len(intersection))
print(*intersection, sep='\n')
```
