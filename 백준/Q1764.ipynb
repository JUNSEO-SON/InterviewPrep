````python
import sys

input=sys.stdin.readline

N,M=map(int,input().split())

never_heard=set([input().strip() for _ in range(N)])
never_seenheard=[]

for _ in range(M):
    never_seen=input().strip()
    if never_seen in never_heard:
        never_seenheard.append(never_seen)

never_seenheard.sort()

print(len(never_seenheard))
print(*never_seenheard,sep='\n')
````
