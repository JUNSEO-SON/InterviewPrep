````python
import sys

stack_lst=[]
count = 1
result = []
n=int(sys.stdin.readline())

for _ in range(n):

    target=int(sys.stdin.readline())

    while (count<=target):
        stack_lst.append(count)
        result.append('+')
        count+=1

    if target==stack_lst[-1]:
            stack_lst.pop()
            result.append('-')
    else:
            print('NO')
            sys.exit()


print(*result, sep='\n')
````

### \#\# Gemini 추천 코드 (다른 제어 흐름 방식)

`sys.exit()` 대신, '플래그(flag)' 변수를 사용하여 성공/실패 상태를 관리하는 방식입니다. 이는 프로그램의 흐름을 갑자기 중단시키지 않아 조금 더 구조적인 스타일입니다.

```python
import sys

n = int(sys.stdin.readline())
stack = []
result_ops = []
count = 1
possible = True # 성공 가능성을 나타내는 플래그

for _ in range(n):
    target = int(sys.stdin.readline())

    while count <= target:
        stack.append(count)
        result_ops.append('+')
        count += 1
    
    if stack[-1] == target:
        stack.pop()
        result_ops.append('-')
    else:
        # 불가능한 경우, 플래그를 False로 바꾸고 즉시 반복을 중단
        possible = False
        break

# for 루프가 끝난 후, 플래그를 확인하여 최종 결과를 출력
if possible:
    print(*result_ops, sep='\n')
else:
    print('NO')
```

### \#\# 1. 알고리즘 측면

  * **평가:** **더할 나위 없이 완벽합니다.** 오름차순으로 숫자를 관리하는 `count` 변수를 두고, 스택을 이용해 목표 수열을 만들어내는 시뮬레이션 방식은 이 문제에 대한 정석적인 최적의 해법입니다. 시간 복잡도, 공간 복잡도 모두 효율적입니다.

### \#\# 3. 또 다른 측면: 프로그램 흐름 제어(Control Flow)

이번 코드에서 논의해 볼 만한 유일한 부분은 **'어떻게 실패를 처리하고 프로그램을 종료할 것인가'** 입니다.

  * **사용자님의 방식: `sys.exit()`**

      * **장점:** 매우 직관적이고 강력합니다. 'NO'가 확정된 순간, 더 이상의 어떤 코드도 실행할 필요 없이 그 자리에서 프로그램을 즉시 끝냅니다. 코딩 테스트처럼 단일 목적을 가진 작은 스크립트에서는 매우 효율적인 방법입니다.
      * **단점:** 프로그램의 흐름이 갑자기 끊기기 때문에, 나중에 함수화하거나 더 큰 프로그램의 일부로 만들 때는 예측하기 어려운 동작을 유발할 수 있습니다.

  * **추천 코드의 방식: 플래그 변수 (`possible`)**

      * **장점:** 프로그램의 흐름이 자연스럽습니다. `for` 루프는 `break`를 통해 정상적으로 종료되고, 루프 바깥에서 '성공했는가, 실패했는가'를 판단하여 그에 맞는 출력을 합니다. 이는 더 구조적이고 일반적인 프로그래밍 스타일에 가깝습니다.
      * **단점:** `possible`이라는 추가 변수를 관리해야 하고, 루프가 끝난 뒤 `if/else` 분기문이 추가로 필요합니다.

**결론:** 두 방식 모두 장단점이 있으며, 정답을 맞히는 데는 아무런 차이가 없습니다. 사용자님의 `sys.exit()` 방식은 코딩 테스트 환경에서 속도와 간결함을 중시하는 훌륭한 선택이며, 플래그 방식은 조금 더 일반적인 소프트웨어 개발 스타일에 가깝다고 이해하시면 됩니다. 이번에도 정말 잘하셨습니다\!
