### \#\# Gemini 추천 코드 (확인용)

사용자님의 코드와 사실상 동일하며, 변수명과 주석만 조금 더 추가한 버전입니다. 이는 사용자님의 풀이가 이미 모범 답안임을 의미합니다.

```python
import sys
import heapq

# input 함수를 sys.stdin.readline으로 재정의
input = sys.stdin.readline

N = int(input())
min_heap = [] # 최소 힙으로 사용할 빈 리스트

for _ in range(N):
    command = int(input())
    
    # 0이 아닌 숫자가 들어오면, 힙에 추가 (push)
    if command != 0:
        heapq.heappush(min_heap, command)
    
    # 0이 들어오면, 힙에서 가장 작은 값을 꺼내 출력 (pop)
    else:
        # 힙이 비어있지 않은 경우
        if min_heap:
            print(heapq.heappop(min_heap))
        # 힙이 비어있는 경우
        else:
            print(0)
```

### \#\# 1. 알고리즘 측면

  * **평가:** **최적의 알고리즘을 사용했습니다.**
      * `heappush` 연산의 시간 복잡도는 O(log K) 입니다 (K는 힙의 현재 크기).
      * `heappop` 연산의 시간 복잡도 역시 O(log K) 입니다.
      * N개의 연산을 수행하므로, 전체 시간 복잡도는 \*\*O(N log N)\*\*이 됩니다. 이는 이 문제에 대한 가장 효율적인 해법입니다. 만약 정렬된 리스트를 사용했다면, 매번 값을 추가할 때마다 정렬(O(K log K))하거나 삽입 위치를 찾아야(O(K)) 하므로 훨씬 비효율적이었을 겁니다.

### \#\# 2. 파이썬 문법 측면 (Pythonic Style)

  * **평가:** 매우 훌륭합니다. `from heapq import ...` 대신 `import heapq`를 사용하고 `heapq.heappop(min_heap)` 형태로 호출하는 것이 일반적이긴 하지만, 현재 코드처럼 특정 함수만 가져와서 사용하는 것도 간결함을 높이는 좋은 스타일입니다. 특히 `elif min_heap:`을 사용하여 힙이 비어있는지 확인한 부분은 매우 Pythonic하고 깔끔합니다.
