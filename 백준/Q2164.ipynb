```python
#    
from collections import deque

def solve_card_problem(n):
    #      
    dq = deque(range(1, n + 1))
    while len(dq) > 1:
        dq.popleft()
        dq.append(dq.popleft())
    #  
    return dq[0]

N = int(input())
result = solve_card_problem(N)
print(result)
```
