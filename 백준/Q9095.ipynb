```python
import sys

input = sys.stdin.readline

# 1. DP   (10)  .
#       n 10  .
dp = [0] * 11

# 2.  () .
dp[1] = 1  # 1
dp[2] = 2  # 1+1, 2
dp[3] = 4  # 1+1+1, 1+2, 2+1, 3

# 3.   DP  .
for i in range(4, 11):
    # i   (i-1)   + (i-2)   + (i-3)   
    dp[i] = dp[i-1] + dp[i-2] + dp[i-3]

# 4.        O(1)   .
T = int(input())
for _ in range(T):
    n = int(input())
    print(dp[n])
```
