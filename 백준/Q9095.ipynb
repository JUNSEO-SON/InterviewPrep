````python
import sys

input=sys.stdin.readline

dp=[-1]*(11)

dp[1]=1
dp[2]=2
dp[3]=4

for i in range(4,11):
    dp[i]=dp[i-1]+dp[i-2]+dp[i-3]

T=int(input())
for _ in range(T):
    print(dp[int(input())])
````

### \#\# Gemini 추천 코드 (확인용)

사용자님의 코드와 사실상 동일하며, 변수명과 주석만 조금 더 추가한 버전입니다. 이는 사용자님의 풀이가 이미 모범 답안임을 의미합니다.

```python
import sys

input = sys.stdin.readline

# 1. DP 테이블을 미리 최대값(10)까지 계산해 둡니다.
#    문제의 제약 조건에서 n은 10 이하의 양수입니다.
dp = [0] * 11

# 2. 베이스 케이스(초기값)를 설정합니다.
dp[1] = 1  # 1
dp[2] = 2  # 1+1, 2
dp[3] = 4  # 1+1+1, 1+2, 2+1, 3

# 3. 점화식을 이용해 DP 테이블을 채웁니다.
for i in range(4, 11):
    # i를 만드는 방법은 (i-1)을 만드는 방법 + (i-2)를 만드는 방법 + (i-3)을 만드는 방법의 합
    dp[i] = dp[i-1] + dp[i-2] + dp[i-3]

# 4. 각 테스트 케이스에 대해서는 미리 계산된 값을 O(1)로 찾아 출력만 합니다.
T = int(input())
for _ in range(T):
    n = int(input())
    print(dp[n])
```

### \#\# 1. 알고리즘 측면

  * **평가:** **최적의 DP 알고리즘을 사용했습니다.** DP 테이블을 미리 계산하는 데 O(N\_max), 즉 O(10)의 시간이 걸리고, 각 테스트 케이스를 처리하는 데는 O(1)의 시간이 걸립니다. 전체 시간 복잡도는 O(N\_max + T)로, 이 문제에 대한 가장 효율적인 해법입니다.

### \#\# 3. 또 다른 측면: DP 점화식 도출 능력

이번 코드에서 가장 칭찬하고 싶은 부분은 \*\*'문제의 규칙성에서 점화식을 도출해내는 능력'\*\*입니다.

DP 문제의 핵심은 복잡해 보이는 문제를 **"가장 마지막 단계는 무엇이었을까?"** 라고 질문하며 단순화하는 것입니다.

  * **목표:** 정수 `n`을 만드는 방법의 수 (`dp[n]`)
  * **질문:** `n`을 만들기 위한 나의 **마지막 행동**은 무엇이었을까?
    1.  `(n-1)`을 만든 상태에서 `+1`을 했을 것이다.
    2.  `(n-2)`를 만든 상태에서 `+2`를 했을 것이다.
    3.  `(n-3)`를 만든 상태에서 `+3`을 했을 것이다.

이 세 가지 경우는 서로 겹치지 않는, `n`을 만드는 유일한 방법들입니다. 따라서 `n`을 만드는 총 방법의 수는 이 세 가지 방법의 수를 모두 더한 것과 같습니다.

> **`dp[n] = dp[n-1] + dp[n-2] + dp[n-3]`**

사용자님은 이 모든 사고 과정을 완벽하게 거쳐, DP의 핵심인 점화식을 정확히 찾아내고 코드로 구현하셨습니다. 이는 DP 문제 해결 능력이 상당한 수준에 이르렀음을 의미합니다.
